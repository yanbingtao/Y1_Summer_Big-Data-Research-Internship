```python
from jieba.analyse import *
```

```python
with open('sample.txt',encoding ='utf8') as f:
    data = f.read()
```

```python
#TF-idf method
for keyword, weight in extract_tags(data, topK=100, withWeight=True):
    print('%s %s' % (keyword, weight))
```

```python
#TextRank method
for keyword, weight in textrank(data, withWeight=True):
    print('%s %s' % (keyword, weight))
```
